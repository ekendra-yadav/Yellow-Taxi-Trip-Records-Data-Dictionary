```python
df['passenger_count'].fillna(df['passenger_count'].median(), inplace=True)
```

```python
df['RatecodeID'].fillna(df['RatecodeID'].mode()[0], inplace=True)
```

```python
df['store_and_fwd_flag'].fillna(df['store_and_fwd_flag'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].map({'N':0, 'Y':1})
```

```python
import matplotlib.pyplot as plt

df['trip_distance'].hist(bins=50)
plt.title("Trip Distance Distribution")
plt.xlabel("Miles")
plt.ylabel("Frequency")
plt.show()

df['total_amount'].hist(bins=50)
plt.title("Total Fare Distribution")
plt.xlabel("Fare Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour

df.groupby('pickup_hour')['total_amount'].mean().plot()
plt.title("Average Fare by Pickup Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Average Fare")
plt.show()
```

```python
df['trip_duration'] = (
    df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
).dt.total_seconds() / 60
```

```python
df = pd.get_dummies(df, columns=['payment_type'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_cols = ['trip_distance', 'trip_duration', 'passenger_count']
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.drop([
'fare_amount',
'tip_amount',
'tolls_amount',
'extra',
'mta_tax',
'improvement_surcharge'
], axis=1, inplace=True)
```
